```python

# Create color palettes
model_palette = {"MLP": "#e8a6d3", "RandomForest": "#a97dc2"}
opt_palette = {"Bayesian": "#92dcb0", "Genetic": "#e8d98a"}

# Create the dashboard
fig = plt.figure(figsize=(16, 12))

# 1. PRIMARY METRICS COMPARISON
ax1 = plt.subplot(2, 2, 1)
metrics = ['MAE', 'R2', 'Pearson_r']
x = np.arange(len(metrics))
width = 0.35

# Assuming you have both MLP and RandomForest in your data
models = df_final['Model'].unique()
for i, model in enumerate(models):
    model_data = df_final[df_final['Model'] == model]
    means = [model_data[metric].mean() for metric in metrics]
    stds = [model_data[metric].std() for metric in metrics]
    
    bars = ax1.bar(x + i*width, means, width, label=model, 
                  yerr=stds, capsize=5, 
                  color=model_palette[model], alpha=0.7)
    
    # Add value labels
    for j, (mean, std) in enumerate(zip(means, stds)):
        ax1.text(x[j] + i*width, mean + std + 0.01, 
                f'{mean:.3f}', ha='center', fontsize=9, fontweight='bold')

ax1.set_xticks(x + width/2)
ax1.set_xticklabels(['MAE\n(Lower better)', 'R\n(Higher better)', 'Pearson r\n(Higher better)'])
ax1.set_ylabel('Score')
ax1.set_title('Primary Performance Metrics by Model', fontsize=12, fontweight='bold')
ax1.legend()
ax1.grid(axis='y', alpha=0.3)

# 2. BAG ANALYSIS: Bias vs Variability
ax2 = plt.subplot(2, 2, 2)

for model in models:
    for optimizer in df_final['Optimization'].unique():
        subset = df_final[(df_final['Model'] == model) & 
                         (df_final['Optimization'] == optimizer)]
        if len(subset) > 0:
            # Calculate mean and std for this group
            mean_bag = subset['BAG_mean'].mean()
            std_bag = subset['BAG_std'].mean()
            
            # Plot point with error bars
            ax2.errorbar(mean_bag, std_bag, 
                        xerr=subset['BAG_mean'].std(), 
                        yerr=subset['BAG_std'].std(),
                        fmt='o', markersize=10, capsize=5,
                        color=model_palette[model],
                        alpha=0.7,
                        label=f'{model} ({optimizer})' if len(subset) > 0 else "")
            
            # Add model name
            ax2.annotate(f'{model[:2]}', (mean_bag, std_bag),
                        xytext=(5, 5), textcoords='offset points',
                        fontsize=9, fontweight='bold')

# Add ideal point lines
ax2.axvline(x=0, color='red', linestyle='--', alpha=0.5, label='Ideal Bias (0)')
ax2.axhline(y=0, color='gray', linestyle=':', alpha=0.3)

ax2.set_xlabel('BAG Mean (Bias) - Closer to 0 is better')
ax2.set_ylabel('BAG Std (Variability) - Lower is better')
ax2.set_title('Bias vs Variability Analysis', fontsize=12, fontweight='bold')
ax2.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
ax2.grid(True, alpha=0.3)

# 3. OPTIMIZER COMPARISON ACROSS MODELS
ax3 = plt.subplot(2, 2, 3)

# Prepare data for grouped bar chart
grouped_data = df_final.groupby(['Model', 'Optimization'])['MAE'].agg(['mean', 'std']).reset_index()

# Create positions for bars
models = grouped_data['Model'].unique()
optimizers = grouped_data['Optimization'].unique()
x = np.arange(len(models))
width = 0.35

for i, optimizer in enumerate(optimizers):
    optimizer_data = grouped_data[grouped_data['Optimization'] == optimizer]
    means = optimizer_data['mean'].values
    stds = optimizer_data['std'].values
    
    ax3.bar(x + i*width, means, width, 
            label=optimizer, yerr=stds, capsize=5,
            color=opt_palette[optimizer], alpha=0.7)
    
    # Add value labels
    for j, (mean, std) in enumerate(zip(means, stds)):
        ax3.text(x[j] + i*width, mean + std + 0.02, 
                f'{mean:.3f}', ha='center', fontsize=9)

ax3.set_xticks(x + width/2)
ax3.set_xticklabels(models)
ax3.set_ylabel('MAE (years)')
ax3.set_title('Optimizer Performance Across Models', fontsize=12, fontweight='bold')
ax3.legend(title='Optimizer')
ax3.grid(axis='y', alpha=0.3)

# 4. CORRELATION HEATMAP
ax4 = plt.subplot(2, 2, 4)

# Select only numeric columns for correlation
numeric_cols = ['MAE', 'R2', 'Pearson_r', 'BAG_mean', 'BAG_std']
corr_matrix = df_final[numeric_cols].corr()

# Create mask for upper triangle
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

# Plot heatmap
sns.heatmap(corr_matrix, mask=mask, annot=True, fmt='.3f', 
           cmap='coolwarm', center=0, square=True,
           cbar_kws={"shrink": 0.8}, ax=ax4,
           annot_kws={"size": 9})

ax4.set_title('Metric Correlations', fontsize=12, fontweight='bold')

plt.suptitle('Model & Optimizer Performance Dashboard', fontsize=16, fontweight='bold', y=0.98)
plt.tight_layout()
plt.savefig(os.path.join(SAVE_DIR, "Model and Optimizer Performance Dashboard.png"))
plt.show()
```

```python
# Create a comprehensive summary table
fig, ax = plt.subplots(figsize=(12, 6))
ax.axis('tight')
ax.axis('off')

# Calculate summary statistics for each model-optimizer combination
summary_data = []

for model in df_final['Model'].unique():
    for optimizer in df_final['Optimization'].unique():
        subset = df_final[(df_final['Model'] == model) & 
                         (df_final['Optimization'] == optimizer)]
        
        if len(subset) > 0:
            summary_data.append([
                f"{model}",
                f"{optimizer}",
                f"{subset['MAE'].mean():.3f}  {subset['MAE'].std():.3f}",
                f"{subset['R2'].mean():.3f}  {subset['R2'].std():.3f}",
                f"{subset['Pearson_r'].mean():.3f}  {subset['Pearson_r'].std():.3f}",
                f"{subset['BAG_mean'].mean():.3f}  {subset['BAG_std'].mean():.3f}"
            ])

# Create table
table = ax.table(cellText=summary_data,
                colLabels=['Model', 'Optimizer', 'MAE', 'R', 'Pearson r', 'BAG (Mean  Std)'],
                cellLoc='center',
                loc='center',
                colWidths=[0.12, 0.12, 0.19, 0.19, 0.19, 0.19])

# Style the table
table.auto_set_font_size(False)
table.set_fontsize(10)

# Color headers
for i in range(len(summary_data[0])):
    table[(0, i)].set_facecolor("#8f78c3")
    table[(0, i)].set_text_props(weight='bold', color='white')

# Alternate row colors
for i in range(1, len(summary_data)+1):
    if i % 2 == 0:
        for j in range(len(summary_data[0])):
            table[(i, j)].set_facecolor('#f2f2f2')

table.scale(1, 1.5)
plt.title('Performance Summary by Model and Optimizer', fontsize=14, fontweight='bold', y=0.98)
plt.tight_layout()
plt.savefig(os.path.join(SAVE_DIR, "Performance Summary Table.png"))
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
models = ['RandomForest (Genetic)', 'RandomForest (Bayesian)', 'MLP (Genetic)', 'MLP (Bayesian)']
seconds = [5.67, 32.46, 1497.91, 1553.12]
minutes = [s/60 for s in seconds]

# Sort for visualization (fastest to slowest)
sorted_indices = np.argsort(seconds)[::-1]  
models = [models[i] for i in sorted_indices]
seconds = [seconds[i] for i in sorted_indices]
minutes = [minutes[i] for i in sorted_indices]

# Create figure with clean aesthetic
fig, ax = plt.subplots(figsize=(12, 6))
fig.patch.set_facecolor('#ffffff')

# Color scheme - professional blues and reds
colors = ["#c3aaec", "#665ba5", "#e79fdc", "#9e5294"]  # Green for RF, Red for MLP

# Create horizontal bars
y_pos = np.arange(len(models))
bars = ax.barh(y_pos, seconds, color=colors, height=0.6, edgecolor='white', linewidth=1)

# Add time labels inside bars (for long bars) and outside (for short bars)
for i, (bar, sec, min_val) in enumerate(zip(bars, seconds, minutes)):
    width = bar.get_width()
    
    # For bars longer than 100 seconds, put text inside
    if width > 100:
        ax.text(width - 50, bar.get_y() + bar.get_height()/2,
                f'{min_val:.1f} min ({sec:.0f} s)',
                va='center', ha='right', fontsize=10, fontweight='bold', color='white')
    else:
        ax.text(width + 20, bar.get_y() + bar.get_height()/2,
                f'{min_val:.1f} min\n{sec:.1f} seconds',
                va='center', ha='left', fontsize=10, fontweight='bold',
                bbox=dict(boxstyle="round,pad=0.2", facecolor="white", alpha=0.9))

# Add speedup annotations
speedup_mlp = seconds[3] / seconds[0]  # MLP vs fastest RF
speedup_rf = seconds[3] / seconds[1]   # MLP vs Bayesian RF

# Draw speedup arrows
ax.annotate(f'{speedup_mlp:.0f}x faster', 
            xy=(seconds[0], 0), xytext=(seconds[3]*0.8, 0.5),
            arrowprops=dict(arrowstyle='->', color='#2c3e50', lw=1.5, alpha=0.7),
            fontsize=11, fontweight='bold', color='#2c3e50',
            ha='center')

ax.annotate(f'{speedup_rf:.0f}x faster', 
            xy=(seconds[1], 1), xytext=(seconds[3]*0.8, 1.5),
            arrowprops=dict(arrowstyle='->', color='#2c3e50', lw=1.5, alpha=0.7),
            fontsize=11, fontweight='bold', color='#2c3e50',
            ha='center')

# Customize axes
ax.set_yticks(y_pos)
ax.set_yticklabels(models, fontsize=11, fontweight='bold')
ax.set_xlabel('Training Time for 20 Iterations (seconds)', fontsize=12, fontweight='bold')
ax.set_title('Training Time Comparison: Random Forest vs MLP', 
             fontsize=14, fontweight='bold', pad=20)

# Add grid for readability
ax.grid(axis='x', alpha=0.2, linestyle='--')
ax.set_axisbelow(True)

# Remove top and right spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_linewidth(0.5)
ax.spines['bottom'].set_linewidth(0.5)

plt.tight_layout()
plt.savefig(os.path.join(SAVE_DIR, 'training_time_clean.png'), dpi=300, bbox_inches='tight', facecolor='white')
plt.show()
```
